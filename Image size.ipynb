## Bibliotecas

```python
import warnings
import numpy as np
from os import listdir
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from os.path import isfile, join
warnings.filterwarnings("ignore")
from skimage import transform
```

## Carregando o Banco de Dados

```python
path = 'DataBase/'
files = [f for f in listdir(path) if isfile(join(path,f))]

images_db = []

for file in files:
    images_db.append(mpimg.imread(join(path,file)))
```

## Uniformizando o tamanho das imagens

```python
height_size =[]
width_size = []

for image in images_db:
    height, width, _ = image.shape
    height_size.append(height)
    width_size.append(width)
    
height_size = np.array(height_size)
width_size = np.array(width_size)
```

```python
fig, axs = plt.subplots(1,2, figsize = (18,6))

plt.subplots_adjust(wspace = .2, hspace = .3)

titles = ['(a)', '(b)']

for ax, title, channel in zip(axs, titles, [height_size, width_size]):
    
    # Plot Data
    
    ax.hist(channel,bins=5)
    
    # Set title
    
    ax.set_title(title, fontsize = 32)
```

```python
fig, axs = plt.subplots(1,2, figsize = (18,6))

plt.subplots_adjust(wspace = .2, hspace = .3)

titles = ['(a)', '(b)']

for ax, title, channel in zip(axs, titles, [height_size[height_size < 500], width_size[width_size < 500]]):
    
    # Plot Data
    
    ax.hist(channel,bins=10)
    
    # Set title
    
    ax.set_title(title, fontsize = 32)
```
